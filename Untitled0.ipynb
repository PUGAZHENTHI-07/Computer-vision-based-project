## Implement Classification Logic

### Subtask:
Inside the function, implement simple logic (e.g., keyword matching, string comparison) to determine which of the predefined categories the input fashion item best fits. The function should return only the category name.


**Reasoning**:
I will implement the classification logic within the `classify_fashion_item` function using keyword matching, converting both the item description and categories to lowercase for case-insensitive comparison, and returning 'Unknown' if no match is found.



```python
def classify_fashion_item(item_description: str) -> str:
    """
    Classifies a fashion item description into one of the predefined categories.

    Args:
        item_description (str): The text description of the fashion item.

    Returns:
        str: The classified category of the fashion item.
    """
    item_description_lower = item_description.lower()

    for category in fashion_categories:
        if category.lower() in item_description_lower:
            return category

    return 'Unknown'

print("Classification function updated successfully.")
```

**Reasoning**:
I need to demonstrate the functionality of the `classify_fashion_item` function by testing it with an example item description, as per the subtask instructions.



```python
example_item = 'a pair of blue denim jeans'
classified_category = classify_fashion_item(example_item)
print(f"Item: '{example_item}'")
print(f"Classified Category: {classified_category}")
```
